## Sentence Generation from Keywords

1. During generation-rule acquisition, generation
rules for each keyword are automatically
acquired.
2. Candidate-text sentences are constructed
during candidate-text construction by applying
the rules acquired in the first
step. Each candidate-text sentence is represented
by a graph or dependency tree.
3. Candidate-text sentences are ranked according
to their scores assigned during evaluation.
The scores are calculated as a
probability estimated by using a keywordproduction
model and a language model
that are trained with a corpus.
4. The candidate-text sentence that maximizes
the score or the candidate-text sentences
whose scores are over a threshold
are selected as output. The system can
also output candidate-text sentences that
are ranked within the top N sentences.


headword: rightmost content word

content word: word whose part-of-speech is a verb, adjective,noun, demonstrative, adverb, conjunction, attribute,

interjection, or undefined word.

function word: everything else, formal nouns and auxiliary verbs “SURU (do)” and “NARU (become)”

```python
body = """<div>
<h3><span class="mw-headline" id="Basics">Basics</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=8" title="Edit section: Basics" data-section="8" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<p><br>
</p>
<dl><dt> Hello.&nbsp;</dt>
<dd> 你好。 Nǐ hǎo.</dd>
<dt> How are you?&nbsp;</dt>
<dd> 你好吗？ Nǐ hǎo ma?</dd>
<dt> Fine, thank you.&nbsp;</dt>
<dd> 很好, 谢谢。 Hěn hǎo, xièxie.</dd>
<dt> May I please ask, what is your name?&nbsp;</dt>
<dd> 请问你叫什么名? Qǐngwèn nǐjiào shěnme míng?</dd>
<dt> What is your name?&nbsp;</dt>
<dd>  你叫什么名字？ Nǐ jiào shénme míngzi?</dd>
<dt> My name is ______ .&nbsp;</dt>
<dd> 我叫 _____ 。 Wǒ jiào ______ . </dd>
<dt> Nice to meet you.&nbsp;</dt>
<dd>  很高兴认识你。 Hěn gāoxìng rènshì nǐ.</dd>
<dt> Please.&nbsp;</dt>
<dd>  请。 Qǐng.</dd>
<dt> Thank you.&nbsp;</dt>
<dd>  谢谢。 Xièxiè.</dd>
<dt> You're welcome.&nbsp;</dt>
<dd> 不客气。 Bú kèqi.</dd>
<dt> Excuse me. (<i>getting attention</i>)&nbsp;</dt>
<dd> 请问 qǐng wèn</dd>
<dt> Excuse me. (<i>begging pardon</i>)&nbsp;</dt>
<dd> 打扰一下。 Dǎrǎo yixià&nbsp;; 麻烦您了, Máfán nín le.</dd>
<dt> Excuse me. (<i>coming through</i>)&nbsp;</dt>
<dd> 对不起 Duìbùqǐ * or * 请让一下 Qǐng ràng yixià</dd>
<dt> I'm sorry.&nbsp;</dt>
<dd> 对不起。 Duìbùqǐ.</dd>
<dt> It's okay. (polite response to "I'm sorry")</dt>
<dd> 没关系 (méiguānxi).</dd>
<dt> Goodbye&nbsp;</dt>
<dd>  再见。 Zàijiàn</dd>
<dt> Goodbye (<i>informal</i>)&nbsp;</dt>
<dd> 拜拜。 Bāi-bāi (Bye bye)</dd>
<dt> I can't speak Chinese.&nbsp;</dt>
<dd>  我不会讲中文。 Wǒ bú huì jiáng zhōngwén.</dd>
<dt> Do you speak English?&nbsp;</dt>
<dd> 你会说英语吗？ Nǐ huì shuō Yīngyǔ ma?</dd>
<dt> Is there someone here who speaks English?&nbsp;</dt>
<dd> 这里有人会说英语吗？ Zhèlĭ yǒu rén hùi shuō Yīngyǔ ma?</dd>
<dt> Help! (in emergencies)</dt>
<dd> 救命！ Jiùmìng!</dd>
<dt> Good morning.&nbsp;</dt>
<dd>  早安。 zǎo ān.</dd>
<dt> Good evening.&nbsp;</dt>
<dd>  晚上好。 Wǎnshàng hǎo.</dd>
<dt> Good night.&nbsp;</dt>
<dd>  晚安。 Wǎn'ān.</dd>
<dt> I don't understand (something heard).&nbsp;</dt>
<dd> 我听不懂。 Wǒ tīng bù dǒng.</dd>
<dt> I don't understand (something read).&nbsp;</dt>
<dd> 我看不懂。 Wǒ kàn bù dǒng.</dd>
<dt> Where is the toilet?&nbsp;</dt>
<dd> 厕所在哪里？ Cèsuǒ zài nǎli?</dd>
<dt> Where is the bathroom（polite)?&nbsp;</dt>
<dd> 洗手间在哪里？ Xǐshǒujiān zài nǎli?</dd></dl>
<h3><span class="mw-headline" id="Problems">Problems</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=9" title="Edit section: Problems" data-section="9" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>

<dl><dt> Leave me alone.&nbsp;</dt>
<dd> 不要打扰我。 (<i>búyào dǎrǎo wǒ</i>)</dd>
<dt> I don't want it! (useful for people who come up trying to sell you something)&nbsp;</dt>
<dd> 我不要 (<i>wǒ búyào!</i>)</dd>
<dt> Don't touch me!&nbsp;</dt>
<dd> 不要碰我！ (<i>búyào pèng wǒ!</i>)</dd>
<dt> I'll call the police.&nbsp;</dt>
<dd> 我会叫警察。 (<i>wǒ yào jiào jǐngchá le</i>) </dd>
<dt> Police!&nbsp;</dt>
<dd> 警察！ (<i>jǐngchá!</i>) or in Mainland China: 公安！(gōng'ān) (either will be understood in Mainland China)</dd>
<dt> Stop! Thief!&nbsp;</dt>
<dd> 住手！小偷！ (<i>zhùshǒu! xiǎotōu!</i>)</dd>
<dt> I need your help.&nbsp;</dt>
<dd> 我需要你的帮助。 (<i>wǒ xūyào nǐde bāngzhù</i>) </dd>
<dt> It's an emergency.&nbsp;</dt>
<dd> 这是紧急情况。 (<i>zhèshì jǐnjí qíngkuàng</i>)</dd>
<dt> I'm lost.&nbsp;</dt>
<dd> 我迷路了。 (<i>wǒ mílù le</i>)</dd>
<dt> I lost my bag.&nbsp;</dt>
<dd> 我丟了手提包。 (<i>wǒ diūle shǒutíbāo</i>)</dd>
<dt> I lost my wallet.&nbsp;</dt>
<dd> 我丟了钱包。 (<i>wǒ diūle qiánbāo</i>)</dd>
<dt> I'm sick.&nbsp;</dt>
<dd> 我生病了。 (<i>wǒ shēngbìng le</i>)</dd>
<dt> I've been injured.&nbsp;</dt>
<dd> 我受伤了。 (<i>wǒ shòushāng le</i>)</dd>
<dt> I need a doctor.&nbsp;</dt>
<dd> 我需要医生。 (<i>wǒ xūyào yīshēng</i>)</dd>
<dt> Can I make a telephone call?&nbsp;</dt>
<dd> 我可以打个电话吗？ (<i>wǒ kěyǐ dǎ ge diànhuà ma?</i>)</dd>
<dt> Can I use your phone?&nbsp;</dt>
<dd>我可以用您的电话吗？(wǒ kěyǐ yòng nín de diànhuà ne?)</dd>
<dt> Can I use your cell phone?&nbsp;</dt>
<dd>我可以用您的手机吗？(wǒ kěyǐ yòng nín de shǒujī ma?)</dd></dl>
<h3><span class="mw-headline" id="Going_to_the_doctor">Going to the doctor</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=10" title="Edit section: Going to the doctor" data-section="10" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> I am sick.&nbsp;</dt>
<dd> 我生病了。 (<i>wǒ shēngbìng le</i>)</dd>
<dt> Painful.&nbsp;</dt>
<dd> 痛。  (<i>tòng</i>)</dd>
<dt> Uncomfortable.&nbsp;</dt>
<dd> 不舒服。 (<i>bù shūfú</i>)</dd>
<dt> Itchy/ticklish.&nbsp;</dt>
<dd> 痒。(<i>yǎng</i>)</dd>
<dt> Sore (In muscle strains).&nbsp;</dt>
<dd> 酸。(<i>suān</i>)</dd>
<dt> Fever.&nbsp;</dt>
<dd> 发烧。 (<i>fāshāo</i>)</dd>
<dt> Cough.&nbsp;</dt>
<dd> 咳嗽。 (<i>késòu</i>)</dd>
<dt> Sneeze.&nbsp;</dt>
<dd> 打喷嚏 (<i>dǎ pēntì</i>)</dd>
<dt> Diarrhoea.&nbsp;</dt>
<dd> 泻肚子/拉肚子 (<i>xiè dùzi/lā dùzi</i>)</dd>
<dt> Running nose.&nbsp;</dt>
<dd> 流鼻涕 (<i>liú bítì</i>)</dd>
<dt> Phlegm.&nbsp;</dt>
<dd> 痰。 (<i>tán</i>)</dd>
<dt> Hands/Arms.&nbsp;</dt>
<dd> 手。 (<i>shǒu</i>)</dd>
<dt> Fingers.&nbsp;</dt>
<dd> 手指。(<i>shǒuzhǐ</i>)</dd>
<dt> Wrist.&nbsp;</dt>
<dd> 手腕。 (<i>shǒuwàn</i>)</dd>
<dt> Shoulder.&nbsp;</dt>
<dd> 肩膀。 (<i>jiānbǎng</i>)</dd>
<dt> Feet.&nbsp;</dt>
<dd> 脚。 (<i>jiǎo</i>)</dd>
<dt> Toes.&nbsp;</dt>
<dd> 脚指。 (<i>jiǎozhǐ</i>)</dd>
<dt> Legs.&nbsp;</dt>
<dd> 腿。 (<i>tuǐ</i>)</dd>
<dt> Nails.&nbsp;</dt>
<dd> 指甲。 (<i>zhǐjiǎ</i>)</dd>
<dt> Body.&nbsp;</dt>
<dd> 身体。 (<i>shēntǐ</i>)</dd>
<dt> Eyes.&nbsp;</dt>
<dd> 眼睛。 (<i>yǎnjīng</i>)</dd>
<dt> Ears.&nbsp;</dt>
<dd> 耳朵。 (<i>ěrduo</i>)</dd>
<dt> Nose.&nbsp;</dt>
<dd> 鼻子。 (<i>bízi</i>)</dd>
<dt> Face.&nbsp;</dt>
<dd> 脸。 (<i>liǎn</i>)</dd>
<dt> Hair.&nbsp;</dt>
<dd> 头发。 (<i>tóufǎ</i>)</dd>
<dt> Head.&nbsp;</dt>
<dd> 头。 (<i>tóu</i>)</dd>
<dt> Neck.&nbsp;</dt>
<dd> 颈项/脖子。 (<i>jǐngxiàng</i>/<i>bózi</i>)</dd>
<dt> Throat.&nbsp;</dt>
<dd> 喉咙。 (<i>hóulóng</i>)</dd>
<dt> Chest.&nbsp;</dt>
<dd> 胸。 (<i>xiōng</i>)</dd>
<dt> Abdomen.&nbsp;</dt>
<dd> 肚子。 (<i>dùzi</i>)</dd>
<dt> Hip/Waist.&nbsp;</dt>
<dd> 腰。 (<i>yāo</i>)</dd>
<dt> Buttocks.&nbsp;</dt>
<dd> 屁股。 (<i>pìgǔ</i>)</dd>
<dt> Back.&nbsp;</dt>
<dd> 背。 (<i>bèi</i>)</dd></dl>
<h3><span class="mw-headline" id="Numbers">Numbers</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=11" title="Edit section: Numbers" data-section="11" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<p>Chinese numbers are very regular. While Arabic numerals have become more common, the Chinese numerals shown below are still used, particularly in informal contexts like markets. The characters in parentheses can be seen as upper case of Chinese numerals. They are generally used in financial contexts, such as writing cheques and printing banknotes.
</p>
<dl><dt> 0&nbsp;</dt>
<dd> 〇 (零) líng</dd>
<dt> 1&nbsp;</dt>
<dd> 一 (壹) yī</dd>
<dt> 2&nbsp;</dt>
<dd> 二 (贰) èr (两 liǎng is used when specifying quantities)</dd>
<dt> 3&nbsp;</dt>
<dd> 三 (叁) sān</dd>
<dt> 4&nbsp;</dt>
<dd> 四 (肆) sì</dd>
<dt> 5&nbsp;</dt>
<dd> 五 (伍) wǔ</dd>
<dt> 6&nbsp;</dt>
<dd> 六 (陆) liù</dd>
<dt> 7&nbsp;</dt>
<dd> 七 (柒) qī</dd>
<dt> 8&nbsp;</dt>
<dd> 八 (捌) bā</dd>
<dt> 9&nbsp;</dt>
<dd> 九 (玖) jiǔ</dd>
<dt> 10&nbsp;</dt>
<dd> 十 (拾) shí</dd>
<dt> 11&nbsp;</dt>
<dd> 十一 shí-yī</dd>
<dt> 12&nbsp;</dt>
<dd> 十二 shí-èr</dd>
<dt> 13&nbsp;</dt>
<dd> 十三 shí-sān</dd>
<dt> 14&nbsp;</dt>
<dd> 十四 shí-sì</dd>
<dt> 15&nbsp;</dt>
<dd> 十五 shí-wǔ</dd>
<dt> 16&nbsp;</dt>
<dd> 十六 shí-liù</dd>
<dt> 17&nbsp;</dt>
<dd> 十七 shí-qī</dd>
<dt> 18&nbsp;</dt>
<dd> 十八 shí-bā</dd>
<dt> 19&nbsp;</dt>
<dd> 十九 shí-jiǔ</dd>
<dt> 20&nbsp;</dt>
<dd> 二十 èr-shí</dd>
<dt> 21&nbsp;</dt>
<dd> 二十一 èr-shí-yī</dd>
<dt> 22&nbsp;</dt>
<dd> 二十二 èr-shí-èr</dd>
<dt> 23&nbsp;</dt>
<dd> 二十三 èr-shí-sān</dd>
<dt> 30&nbsp;</dt>
<dd> 三十 sān-shí</dd>
<dt> 40&nbsp;</dt>
<dd> 四十 sì-shí</dd>
<dt> 50&nbsp;</dt>
<dd> 五十 wǔ-shí</dd>
<dt> 60&nbsp;</dt>
<dd> 六十 liù-shí</dd>
<dt> 70&nbsp;</dt>
<dd> 七十 qī-shí</dd>
<dt> 80&nbsp;</dt>
<dd> 八十 bā-shí</dd>
<dt> 90&nbsp;</dt>
<dd> 九十 jiǔ-shí</dd></dl>
<p>For numbers above 100, any "gaps" must be filled in with 〇 <i>líng</i>, as eg. 一百一 <i>yībǎiyī</i> would otherwise be taken as shorthand for "110". A single unit of tens may be written and pronounced either 一十 <i>yīshí</i> or just 十 <i>shí</i>.
</p>
<dl><dt> 100&nbsp;</dt>
<dd> 一百 (壹佰) yī-bǎi</dd>
<dt> 101&nbsp;</dt>
<dd> 一百〇一 (壹佰零壹) yī-bǎi-líng-yī</dd>
<dt> 110&nbsp;</dt>
<dd> 一百一十 (壹佰壹拾) yī-bǎi-yī-shí</dd>
<dt> 111&nbsp;</dt>
<dd> 一百一十一 yī-bǎi-yī-shí-yī</dd>
<dt> 200&nbsp;</dt>
<dd> 二百 èr-bǎi or 两百：liǎng-bǎi</dd>
<dt> 300&nbsp;</dt>
<dd> 三百 sān-bǎi</dd>
<dt> 500&nbsp;</dt>
<dd> 五百 wǔ-bǎi</dd>
<dt> 1000&nbsp;</dt>
<dd> 一千 (壹仟) yī-qiān</dd>
<dt> 2000&nbsp;</dt>
<dd> 二千 èr-qiān</dd>
<dd> <i>or</i> 两千 liǎng-qiān</dd></dl>
<p>Numbers starting from 10,000 are grouped by in units of four digits starting with 万 <i>wàn</i> (ten thousand). "One million" in Chinese is thus "hundred ten-thousands" (一百万).
</p>
<dl><dt> 10,000&nbsp;</dt>
<dd> 一万 (壹萬) yī-wàn</dd>
<dt> 10,001&nbsp;</dt>
<dd> 一万〇一 yī-wàn-líng-yī</dd>
<dt> 10,002&nbsp;</dt>
<dd> 一万〇二 yī-wàn-líng-èr</dd>
<dt> 20,000&nbsp;</dt>
<dd> 二万 èr-wàn</dd>
<dt> 50,000&nbsp;</dt>
<dd> 五万 wǔ-wàn</dd>
<dt> 100,000&nbsp;</dt>
<dd> 十万 shí-wàn</dd>
<dt> 200,000&nbsp;</dt>
<dd> 二十万 èr-shí-wàn</dd>
<dt> 1,000,000&nbsp;</dt>
<dd> 一百万 yī-bǎi-wàn</dd>
<dt> 10,000,000&nbsp;</dt>
<dd> 一千万 yī-qiān-wàn</dd>
<dt> 100,000,000&nbsp;</dt>
<dd> 一亿 (壹億) yī-yì</dd>
<dt> 1,000,000,000,000&nbsp;</dt>
<dd> 一兆 yī-zhào</dd></dl>

<dl><dt> half&nbsp;</dt>
<dd> 半 bàn</dd>
<dt> less than&nbsp;</dt>
<dd> 少於 shǎoyú</dd>
<dt> more than&nbsp;</dt>
<dd> 多於 duōyú</dd>
<dt> more&nbsp;</dt>
<dd> 更 gèng</dd></dl>
<h3><span class="mw-headline" id="Time">Time</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=12" title="Edit section: Time" data-section="12" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> now&nbsp;</dt>
<dd> 现在 xiànzài</dd>
<dt> later&nbsp;</dt>
<dd> 以后 <i>or</i> 稍后， yǐhòu <i>or</i> shāohòu</dd>
<dt> before&nbsp;</dt>
<dd> 以前, yǐqián</dd>
<dt> morning&nbsp;</dt>
<dd> 早上, zǎoshàng</dd>
<dt> noon</dt>
<dd> 中午, zhōngwǔ</dd>
<dt> afternoon&nbsp;</dt>
<dd> 下午, xiàwǔ</dd>
<dt> evening/night&nbsp;</dt>
<dd> 晚上, wǎnshàng</dd>
<dt> midnight</dt>
<dd> 半夜 bànyè or 午夜 (wǔyè)</dd></dl>
<h4><span class="mw-headline" id="Clock_time">Clock time</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=13" title="Edit section: Clock time" data-section="13" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>
<dl><dt> What time is it?&nbsp;</dt>
<dd> 现在几点？ Xiànzài jǐ diǎn?</dd>
<dt> It is nine in the morning.&nbsp;</dt>
<dd> 早上9点钟。 Zǎoshàng jǐu diǎn zhōng.</dd>
<dt> Three-thirty PM.&nbsp;</dt>
<dd> 下午3点半. Xiàwǔ sān diǎn bàn.</dd>
<dt> 3:38 PM&nbsp;</dt>
<dd> 下午3点38分 Xiàwǔ sāndiǎn sānshíbā fēn.</dd></dl>
<h4><span class="mw-headline" id="Duration">Duration</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=14" title="Edit section: Duration" data-section="14" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>
<dl><dt> _____ minute(s)&nbsp;</dt>
<dd> _____ 分钟 fēnzhōng</dd>
<dt> _____ hour(s)&nbsp;</dt>
<dd> _____ 小时 xiǎoshí</dd>
<dt> _____ day(s)&nbsp;</dt>
<dd> _____ 天 tiān</dd>
<dt> _____ week(s)&nbsp;</dt>
<dd> _____ 星期 xīngqī</dd>
<dt> _____ month(s)&nbsp;</dt>
<dd> _____ 月 yùe</dd>
<dt> _____ year(s)&nbsp;</dt>
<dd> _____ 年 nián</dd></dl>
<h4><span class="mw-headline" id="Days">Days</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=15" title="Edit section: Days" data-section="15" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>
<dl><dt> today&nbsp;</dt>
<dd> 今天 jīntiān</dd>
<dt> yesterday&nbsp;</dt>
<dd> 昨天 zuótiān</dd>
<dt> the day before yesterday</dt>
<dd> 前天qiăntiān</dd>
<dt> tomorrow&nbsp;</dt>
<dd> 明天 míngtiān</dd>
<dt> the day after tomorrow</dt>
<dd> 后天 hòutiān</dd></dl>
<dl><dt> this week&nbsp;</dt>
<dd> 这个星期 zhè ge xīngqī</dd>
<dt> last week&nbsp;</dt>
<dd> 上个星期 shàng ge xīngqī</dd>
<dt> next week&nbsp;</dt>
<dd> 下个星期 xià ge xīngqī</dd></dl>
<p>Weekdays in Chinese are easy: starting with 1 for Monday, just add the number after 星期 xīngqī. In <a href="/en/Taiwan" title="Taiwan">Taiwan</a>, 星期 is pronounced xīngqí (second tone on the second syllable).
</p>
<dl><dt> Sunday&nbsp;</dt>
<dd> 星期天 xīngqī tiān <i>or</i> xīngqī rì (星期日)</dd>
<dt> Monday&nbsp;</dt>
<dd> 星期一 xīngqī yī</dd>
<dt> Tuesday&nbsp;</dt>
<dd> 星期二 xīngqī èr</dd>
<dt> Wednesday&nbsp;</dt>
<dd> 星期三 xīngqī sān</dd>
<dt> Thursday&nbsp;</dt>
<dd> 星期四 xīngqī sì</dd>
<dt> Friday&nbsp;</dt>
<dd> 星期五 xīngqī wǔ</dd>
<dt> Saturday&nbsp;</dt>
<dd> 星期六 xīngqī liù</dd></dl>
<p>星期 can also be replaced with 礼拜 lǐbài and occasionally 周 zhōu. eg. 星期三=礼拜三=周三/ xīngqī sān = lǐbài sān = zhōu sān
</p>
<h4><span class="mw-headline" id="Months">Months</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=16" title="Edit section: Months" data-section="16" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>
<p>Months in Chinese are also easy: starting with 1 for January, just add the number before 月 yuè.
</p>
<dl><dt> January&nbsp;</dt>
<dd> 一月, yī yuè</dd>
<dt> February&nbsp;</dt>
<dd> 二月, èr yuè</dd>
<dt> March&nbsp;</dt>
<dd> 三月, sān yuè</dd>
<dt> April&nbsp;</dt>
<dd> 四月, sì yuè</dd>
<dt> May&nbsp;</dt>
<dd> 五月, wŭ yuè</dd>
<dt> June&nbsp;</dt>
<dd> 六月, liù yuè</dd>
<dt> July&nbsp;</dt>
<dd> 七月, qī yuè</dd>
<dt> August&nbsp;</dt>
<dd> 八月, bā yuè</dd>
<dt> September&nbsp;</dt>
<dd> 九月, jiŭ yuè</dd>
<dt> October&nbsp;</dt>
<dd> 十月, shí yuè</dd>
<dt> November&nbsp;</dt>
<dd> 十一月, shí yī yuè</dd>
<dt> December&nbsp;</dt>
<dd> 十二月, shí èr yuè</dd></dl>
<p>From January to December, you just need to use this pattern: number (1-12) + yuè.
</p>
<h4><span class="mw-headline" id="Writing_Dates">Writing Dates</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=17" title="Edit section: Writing Dates" data-section="17" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>

<p><br>
When writing the date, you name the month <i>(number (1-12) + yuè)</i>, before inserting the day <i>(number (1-31) + 日(rì)/ 号(hào) )</i>. Note that the usage of 号(hào), which is more often used in spoken language, is more colloquial than that of 日(rì), which is more often used in written documents.
</p>
<dl><dt>6th January</dt>
<dd> 一月六号 (yī yuè liù hào) or 一月六日 (yī yuè liù rì)</dd>
<dt>25th December</dt>
<dd> 十二月二十五号 (shí-èr yuè èr-shí-wǔ hào)</dd></dl>
<h3><span class="mw-headline" id="Colours">Colours</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=18" title="Edit section: Colours" data-section="18" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> black&nbsp;</dt>
<dd> 黑色 hēi sè</dd>
<dt> white&nbsp;</dt>
<dd> 白色 bái sè</dd>
<dt> grey&nbsp;</dt>
<dd> 灰色 huī sè</dd>
<dt> red&nbsp;</dt>
<dd> 红色 hóng sè</dd>
<dt> blue&nbsp;</dt>
<dd> 蓝色 lán sè</dd>
<dt> yellow&nbsp;</dt>
<dd> 黄色 huáng sè</dd>
<dt> green&nbsp;</dt>
<dd> 绿色  lǜ sè </dd>
<dt> orange&nbsp;</dt>
<dd> 橙色 chéng sè</dd>
<dt> purple&nbsp;</dt>
<dd> 紫色 zǐ sè</dd>
<dt> brown&nbsp;</dt>
<dd> 褐色 he sè,  棕色 zōng sè,  </dd>
<dt> gold&nbsp;</dt>
<dd> 金色 jīn se </dd></dl>
<dl><dt> Do you have it in another colour? &nbsp;</dt>
<dd> 你们有没有另外颜色?  nǐmen yǒu méiyǒu lìngwài yánsè&nbsp;?</dd></dl>

<h3><span class="mw-headline" id="Transportation">Transportation</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=19" title="Edit section: Transportation" data-section="19" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<h4><span class="mw-headline" id="Bus_and_Train">Bus and Train</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=20" title="Edit section: Bus and Train" data-section="20" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>

<dl><dt> How much is a ticket to _____?&nbsp;</dt>
<dd> 去______的票多少钱? qù _____ de piào duō shǎo qián?</dd>
<dt> Do you go to ___(the central station)?&nbsp;</dt>
<dd> 去不去___(火车站)? qù bù qù ___ (huǒ chē zhàn)?</dd></dl>
<h4><span class="mw-headline" id="Directions">Directions</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=21" title="Edit section: Directions" data-section="21" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>
<dl><dt> How do I get to _____&nbsp;?&nbsp;</dt>
<dd> 怎么去_____ zěnme  qù _____? </dd>
<dt> ...the train station?&nbsp;</dt>
<dd> ...火车站? ...huǒchē zhàn?</dd>
<dt> ...the bus station?&nbsp;</dt>
<dd> ...汽车总站? ...qìchē  zǒngzhàn?</dd>
<dt> ...the airport?&nbsp;</dt>
<dd> ...机场? ...jī chǎng?</dd></dl>
<dl><dt> street&nbsp;</dt>
<dd> 街 jiē;</dd>
<dt> road&nbsp;</dt>
<dd> 路 lù</dd>
<dt> avenue&nbsp;</dt>
<dd> 大道 dàjiē</dd></dl>
<dl><dt> Turn left.&nbsp;</dt>
<dd> 左边转弯 zuǒbiān zhuǎnwān/左拐zuǒguǎi</dd>
<dt> Turn right.&nbsp;</dt>
<dd> 右边转弯 yòubiān zhuǎnwān/右拐yòuguăi</dd>
<dt> Go straight</dt>
<dd> 一直走 yìzhízŏu  </dd>
<dt> I've reached my destination</dt>
<dd> 到了dàole                                    </dd>
<dt> U-turn</dt>
<dd> 掉头 diàotóu </dd>
<dt> Taxi driver</dt>
<dd> 司机 sījī</dd>
<dt> Please use the meter machine</dt>
<dd>  请打表 qǐng dǎbiǎo                                        </dd>
<dt> Please turn up the aircon/heater</dt>
<dd> 请把空调开大一点。 qǐng kōngtiáo kāi dàdiǎn</dd></dl>
<dl><dt> left&nbsp;</dt>
<dd> 左边 zuǒbiān</dd>
<dt> right&nbsp;</dt>
<dd> 右边 yòubiān</dd>
<dt> straight ahead&nbsp;</dt>
<dd> 往前走 wǎngqián zǒu</dd>
<dt> north&nbsp;</dt>
<dd> 北 bĕi </dd>
<dt> south&nbsp;</dt>
<dd> 南 nán </dd>
<dt> east&nbsp;</dt>
<dd> 东 dōng</dd>
<dt> west&nbsp;</dt>
<dd> 西 xī</dd></dl>
<h4><span class="mw-headline" id="Taxi">Taxi</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=22" title="Edit section: Taxi" data-section="22" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h4>
<dl><dt> Taxi  出租车 chū zū chē</dt>
<dt> Take me to _____, please.&nbsp;</dt>
<dd> 请您开到_____。 qǐng nín kāidào _____。</dd></dl>
<h3><span class="mw-headline" id="Lodging">Lodging</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=23" title="Edit section: Lodging" data-section="23" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>

<p><br>
</p>
<dl><dt> Do you have any rooms available?&nbsp;</dt>
<dd> 你们有房间吗？ Nǐmen yǒu fángjiān ma?</dd>
<dt> Does the room come with...&nbsp;</dt>
<dd> 有没有... Yǒu méiyǒu...</dd>
<dt> ...bedsheets?&nbsp;</dt>
<dd> ...床单？ ...chuángdān?</dd>
<dt> ...a bathroom?&nbsp;</dt>
<dd> ...浴室？ ...yùshì?</dd>
<dt> ...a telephone?&nbsp;</dt>
<dd> ...电话？ ...diànhuà?</dd>
<dt> ...a TV?&nbsp;</dt>
<dd> ...电视？ ...diànshì&nbsp;?</dd>
<dt> I will stay for _____ night(s).&nbsp;</dt>
<dd> 我打算住_____夜。 Wǒ dǎsuàn zhù _____ yè.</dd>
<dt> Do you have a safe?&nbsp;</dt>
<dd> 你们有没有保险箱？ Nǐmen yǒu méiyǒu bǎoxiǎn xiāng?</dd>
<dt> Can you wake me at _____?&nbsp;</dt>
<dd> 请明天早上_____叫醒我。 Qǐng míngtiān zǎoshàng  _____ jiàoxǐng wǒ.</dd>
<dt> I want to check out.&nbsp;</dt>
<dd> 我现在要走了。 Wǒ xiànzài yào zǒu le.</dd></dl>
<h3><span class="mw-headline" id="Money">Money</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=24" title="Edit section: Money" data-section="24" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> pay</dt>
<dd> 付 fù</dd>
<dt> cash</dt>
<dd> 现钱 xiàn qián</dd>
<dt> credit card</dt>
<dd> 信用卡 xìn yòng kǎ</dd>
<dt> check</dt>
<dd> 支票　zhīpiào</dd></dl>
<h3><span class="mw-headline" id="Eating">Eating</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=25" title="Edit section: Eating" data-section="25" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>

<dl><dt> Can I look at the menu, please?&nbsp;</dt>
<dd> 请给我看看菜单. qǐng gěi wǒ kànkan Càidān.</dd>
<dt> Do you have an English menu?&nbsp;</dt>
<dd> 你有没有英文菜单？ nǐ yŏu méi yǒu yīngwén Càidān?</dd></dl>
<p>(Listen for...
Yes, we have one.&nbsp;: 有 yǒu - No, we don't.&nbsp;: 没有 méi yǒu)
</p>
<dl><dt> I'm a vegetarian&nbsp;</dt>
<dd> 我吃素的  wǒ chī sù de </dd>
<dt> breakfast&nbsp;</dt>
<dd> 早饭 zǎofàn or 早餐 zǎocān</dd>
<dt> lunch&nbsp;</dt>
<dd> 午饭 wǔfàn  <i>or</i> 中饭 zhōngfàn or 午餐 wǔcān</dd>
<dt> supper&nbsp;</dt>
<dd> 晚饭 wǎnfàn or 晚餐 wǎncān</dd>
<dt> beef&nbsp;</dt>
<dd> 牛肉  niúròu </dd>
<dt> pork</dt>
<dd> 猪肉  zhūròu，or sometimes simply '肉' ròu.</dd>
<dt> mutton</dt>
<dd> 羊肉  yángròu</dd>
<dt> chicken</dt>
<dd> 鸡  jī</dd>
<dt> fish</dt>
<dd> 鱼  yú</dd>
<dt> vegetable</dt>
<dd> 蔬菜 shūcài</dd>
<dt> cheese&nbsp;</dt>
<dd> 奶酪  nǎilào</dd>
<dt> eggs&nbsp;</dt>
<dd> 鸡蛋 jīdàn</dd>
<dt> bread&nbsp;</dt>
<dd> 面包  miànbāo </dd>
<dt> noodles&nbsp;</dt>
<dd> 面条 miàntiáo </dd>
<dt> fried rice</dt>
<dd> 炒饭  chǎofàn</dd>
<dt> dumpling</dt>
<dd> 饺子  jiǎozi</dd>
<dt> rice&nbsp;</dt>
<dd> 米饭 mĭfàn  </dd>
<dt> soup&nbsp;</dt>
<dd> 汤 tāng</dd>
<dt> coffee&nbsp;</dt>
<dd> 咖啡 kāfēi</dd>
<dd> black coffee: 黑咖啡 hēi kāfēi</dd>
<dt> milk</dt>
<dd> 牛奶 niúnǎi</dd>
<dt> sugar</dt>
<dd> 糖 táng</dd>
<dt> tea (<i>drink</i>)&nbsp;</dt>
<dd> 茶 chá</dd>
<dt> green tea</dt>
<dd> 绿茶 lǜ chá</dd>
<dt> scented tea</dt>
<dd> 花茶 huāchá</dd>
<dt> black tea</dt>
<dd> 红茶 hóngchá</dd>
<dt> juice&nbsp;</dt>
<dd> (水)果汁  (shuǐ)guǒzhī, literally 'fruit juice'.</dd>
<dt> water&nbsp;</dt>
<dd> 水 shuĭ</dd>
<dt> natural mineral water</dt>
<dd> 矿泉水 kuàngquán shuǐ</dd>
<dt> beer&nbsp;</dt>
<dd> 啤酒 píjiŭ</dd>
<dt> red/white wine&nbsp;</dt>
<dd> 红/白 葡萄 酒 hóng/bái pútáo jiŭ</dd>
<dt> It was delicious.&nbsp;</dt>
<dd> 很好吃。 hěn hǎochī.</dd>
<dt> The check, please.&nbsp;</dt>
<dd> 请结帐。 qǐng jiézhàng.</dd></dl>
<h3><span class="mw-headline" id="Bars">Bars</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=26" title="Edit section: Bars" data-section="26" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> Do you serve alcohol?&nbsp;</dt>
<dd> 卖不卖酒？ (<i> mài búmài jiǔ?</i>)</dd>
<dt> Is there table service?&nbsp;</dt>
<dd> 有没有餐桌服务？ (<i>yǒu méiyǒu cānzhuō fúwù?</i>)</dd>
<dt> A beer/two beers, please.&nbsp;</dt>
<dd> 请给我一杯/两杯啤酒。 (<i>qǐng gěiwǒ yìbēi/liǎngbēi píjiǔ</i>)</dd>
<dt> A glass of red/white wine, please.&nbsp;</dt>
<dd> 请给我一杯红/白酒。 (<i>qǐng gěi wǒ yìbēi hóng/bái jiǔ</i>)</dd>
<dt> A pint, please.&nbsp;</dt>
<dd> 请给我一品脱。 (<i>qǐng gěi wǒ yìpǐntuō</i>)</dd>
<dt> A bottle, please.&nbsp;</dt>
<dd> 请给我一瓶。 (<i>qǐng gěi wǒ yìpíng</i>)</dd>
<dt> _____ (<i>hard liquor</i>) and _____ (<i>mixer</i>), please.&nbsp;</dt>
<dd> 请给我_____和_____。 (<i>qǐng gěi wǒ _____ hé _____</i>)</dd>
<dt> whiskey&nbsp;</dt>
<dd> 威士忌 (<i>wēishìjì</i>)</dd>
<dt> vodka&nbsp;</dt>
<dd> 伏特加 (<i>fútèjiā</i>)</dd>
<dt> rum&nbsp;</dt>
<dd> 朗姆酒 (<i>lángmǔ jiǔ</i>)</dd>
<dt> water&nbsp;</dt>
<dd> 水 (<i>shuǐ</i>)</dd>
<dt> mineral spring (i.e. bottled) water&nbsp;</dt>
<dd> 矿泉水 (kuàngquánshuǐ)</dd>
<dt> boiled water</dt>
<dd> 开水 (kāishuǐ)</dd>
<dt> club soda&nbsp;</dt>
<dd> 苏打水 (sūdǎshuǐ)</dd>
<dt> tonic water&nbsp;</dt>
<dd> 通宁水 (tōngníng shuǐ)/汤力水 (tānglì shuǐ)</dd>
<dt> orange juice&nbsp;</dt>
<dd> 柳橙汁 （liǔchéng zhī)</dd>
<dt> Coke (<i>soda</i>)&nbsp;</dt>
<dd> 可乐 (<i>kělè</i>)</dd>
<dt> Do you have any bar snacks?&nbsp;</dt>
<dd> 有没有吧台小吃？ (<i>yǒu méiyǒu bātái xiaochi?</i>)</dd>
<dt> One more, please.&nbsp;</dt>
<dd> 请再给我一个。 (qǐng zài gěi wǒ yígè')</dd>
<dt> Another round, please.&nbsp;</dt>
<dd> 请再来一轮。 (qǐng zàilái yìlún)</dd>
<dt> When is closing time?&nbsp;</dt>
<dd> 几点打烊/关门？ (jǐdiǎn dǎyáng/guānmén?)</dd>
<dt> Where is the toilet?&nbsp;</dt>
<dd> 厕所在哪里？ (<i>cèsuǒ zài nǎli?</i>)</dd>
<dt> Where is the washingroom?&nbsp;</dt>
<dd> 洗手间在哪儿？（<i>xǐshǒujiān zài nǎr?</i>）</dd></dl>
<h3><span class="mw-headline" id="Shopping">Shopping</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=27" title="Edit section: Shopping" data-section="27" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> Do you have this in my size?&nbsp;</dt>
<dd> 有没有我的尺寸？ (<i>yǒu méiyǒu wǒde chǐcùn?</i>)</dd>
<dt> How much is this?&nbsp;</dt>
<dd> 多少钱？ (<i>duōshǎo qián?</i>)</dd>
<dt> That's too expensive.&nbsp;</dt>
<dd> 太贵了。 (<i>tài guì le</i>)</dd>
<dt> Would you take _____?&nbsp;</dt>
<dd> _____元可以吗？ (<i>_____ yuán kěyǐ ma?</i>)</dd>
<dt> expensive&nbsp;</dt>
<dd> 贵 (<i>guì</i>)</dd>
<dt> cheap&nbsp;</dt>
<dd> 便宜 (<i>piányi</i>)</dd>
<dt> I can't afford it.&nbsp;</dt>
<dd> 我付不起。 (<i>wǒ fù bù qǐ</i>)</dd>
<dt> I don't want it.&nbsp;</dt>
<dd> 我不要。 (<i>wǒ bú yào</i>)</dd>
<dt> You're cheating me.&nbsp;</dt>
<dd> 你在骗我。 (<i>nǐ zài piàn wǒ</i>) <b>Use with caution!</b></dd>
<dt> I'm not interested.&nbsp;</dt>
<dd> 我没有兴趣。 (<i>wǒ méiyǒu xìngqù</i>)</dd>
<dt> OK, I'll take it.&nbsp;</dt>
<dd> 我要买这个。 (<i>wǒ yào mǎi zhège</i>)</dd>
<dt> Please provide me with a carrier-bag.&nbsp;</dt>
<dd> 请给我袋子。 (<i>qǐng gěi wǒ dàizi</i>)</dd>
<dt> Do you ship (overseas)?&nbsp;</dt>
<dd> 可以邮寄到海外吗？ (<i>kěyǐ yóujì dào hǎiwài ma?</i>)</dd>
<dt> I need...&nbsp;</dt>
<dd> 我要_____ (<i>wǒ yào _____</i>)</dd>
<dt> ...toothpaste.&nbsp;</dt>
<dd> 牙膏 (<i>yágāo</i>)</dd>
<dt> ...a toothbrush.&nbsp;</dt>
<dd> 牙刷 (<i>yáshuā</i>)</dd>
<dt> ...tampons.&nbsp;</dt>
<dd> 卫生棉条 (<i>wèishēng miántiáo</i>)</dd>
<dt> ...soap.&nbsp;</dt>
<dd> 肥皂 (<i>féizào</i>)</dd>
<dt> ...shampoo.&nbsp;</dt>
<dd> 洗发精 (<i>xǐfàjīng</i>) </dd>
<dt> ...pain reliever. (<i>e.g., aspirin or ibuprofen</i>)&nbsp;</dt>
<dd> 镇痛剂 (<i>zhèntòngjì</i>)</dd>
<dt> ...cold medicine.&nbsp;</dt>
<dd> 感冒药 (<i>gǎnmào yào</i>)</dd>
<dt> ...stomach medicine.&nbsp;</dt>
<dd> 胃肠药 (<i>wèicháng yào</i>)</dd>
<dt> ...a razor.&nbsp;</dt>
<dd> 剃刀 (<i>tìdāo</i>)</dd>
<dt> ...an umbrella.&nbsp;</dt>
<dd> 雨伞 (<i>yǔsǎn</i>)</dd>
<dt> ...sunblock lotion.&nbsp;</dt>
<dd> 防晒油 (<i>fángshàiyóu</i>)</dd>
<dt> ...a postcard.&nbsp;</dt>
<dd> 明信片 (<i>míngxìnpiàn</i>)</dd>
<dt> ...postage stamps.&nbsp;</dt>
<dd> 邮票 (<i>yóupiào</i>)</dd>
<dt> ...batteries.&nbsp;</dt>
<dd> 电池 (<i>diànchí</i>)</dd>
<dt> ...writing paper.&nbsp;</dt>
<dd> 纸 (<i>zhǐ</i>)</dd>
<dt> ...a pen.&nbsp;</dt>
<dd> 笔 (<i>bǐ</i>)</dd>
<dt> ...a pencil.&nbsp;</dt>
<dd> 铅笔 (<i>qiānbǐ</i>)</dd>
<dt> ...glasses.&nbsp;</dt>
<dd> 眼镜 (<i>yǎnjìng</i>)</dd>
<dt> ...English-language books.&nbsp;</dt>
<dd> 英文书 (<i>Yīngwén shū</i>)</dd>
<dt> ...English-language magazines.&nbsp;</dt>
<dd> 英文杂志 (<i>Yīngwén zázhì</i>)</dd>
<dt> ...an English-language newspaper.&nbsp;</dt>
<dd> 英文报纸 (<i>Yīngwén bàozhǐ</i>)</dd>
<dt> ...a Chinese-English dictionary.&nbsp;</dt>
<dd> 汉英词典 (<i>Hàn-Yīng cídiǎn</i>)</dd>
<dt> ...an English-Chinese dictionary.&nbsp;</dt>
<dd> 英汉词典 (<i>Yīng-Hàn cídiǎn</i>)</dd></dl>
<h3><span class="mw-headline" id="Driving">Driving</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=28" title="Edit section: Driving" data-section="28" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> I want to rent a car.&nbsp;</dt>
<dd> 我想要租车。 (<i>wǒ xiǎngyào zūchē</i>)</dd>
<dt> Can I get insurance?&nbsp;</dt>
<dd> 我可以买保险吗？ (<i>wǒ kěyǐ mǎi bǎoxiǎn ma?</i>)</dd>
<dt> stop (<i>on a street sign</i>)&nbsp;</dt>
<dd> 停 (<i>tíng</i>)</dd>
<dt> one way&nbsp;</dt>
<dd> 单行道 (<i>dānxíngdào</i>)</dd>
<dt> yield&nbsp;</dt>
<dd> 让路 (<i>rànglù</i>)</dd>
<dt> no parking&nbsp;</dt>
<dd> 禁止停车 (<doi>jìnzhǐ tíngchē</i>)</dd>
<dt> speed limit&nbsp;</dt>
<dd> 速度限制 (<i>sùdù xiànzhì</i>)</dd>
<dt> gas (<i>petrol</i>) station&nbsp;</dt>
<dd> 加油站 (<i>jiāyóuzhàn</i>)</dd>
<dt> petrol&nbsp;</dt>
<dd> 汽油 (<i>qìyóu</i>)</dd>
<dt> diesel&nbsp;</dt>
<dd> 柴油 (<i>cháiyóu</i>)</dd></dl>
<h3><span class="mw-headline" id="Authority">Authority</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=29" title="Edit section: Authority" data-section="29" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>
<dl><dt> I haven't done anything wrong.&nbsp;</dt>
<dd> 我没有做错事。 (<i>wǒ méiyǒu zuòcuò shì</i>)</dd>
<dt> It was a misunderstanding.&nbsp;</dt>
<dd> 这是误会。 (<i>zhè shì wùhuì</i>)</dd>
<dt> Where are you taking me?&nbsp;</dt>
<dd> 你带我去哪里？ (<i>nǐ dài wǒ qù nǎlǐ?</i>)</dd>
<dt> Am I under arrest?&nbsp;</dt>
<dd> 我被捕了吗？ (<i>wǒ bèibǔle ma?</i>)</dd>
<dt> I am an American/Australian/British/Canadian citizen.&nbsp;</dt>
<dd> 我是 美国/澳洲/英国/加拿大 公民。 (<i>wǒ shì měiguó/àozhōu/yīngguó/jiānádà gōngmín</i>)</dd>
<dt> I want to talk to the American/Australian/British/Canadian embassy/consulate.&nbsp;</dt>
<dd> 我希望跟 美国/澳洲/英国/加拿大 的 大使馆/领事馆 联系。 (<i>wǒ xīwàng gēn měiguó/àozhōu/yīngguó/jiānádà de dàshǐguǎn/lǐngshìguǎn liánxì</i>)</dd>
<dt> I want to talk to a lawyer.&nbsp;</dt>
<dd> 我希望跟律师联系。 (<i>wǒ xīwàng gēn lǜshī liánxì</i>)</dd>
<dt> Can I just pay a fine now?&nbsp;</dt>
<dd> 我可以支付罚款吗？ (<i>wǒ kěyǐ zhī fù fákuǎn ma?</i>)</dd></dl>
<h3><span class="mw-headline" id="Telephone_and_the_Internet">Telephone and the Internet</span><span><a href="/wiki/en/index.php?title=Chinese_phrasebook&amp;action=edit&amp;section=30" title="Edit section: Telephone and the Internet" data-section="30" class="mw-ui-icon mw-ui-icon-element mw-ui-icon-edit-enabled edit-page">Edit</a></span></h3>

<p><br>
</p>
<dl><dt> Can I make international calls here?&nbsp;</dt>
<dd> 可以打国际电话吗? (<i>kěyǐ dǎ guójì diànhuà ma?</i>)</dd>
<dt> How much is it to America/Australia/Britain/Canada?&nbsp;</dt>
<dd> 打给 美国/澳洲/英国/加拿大 是多少钱? (<i>dǎgěi měiguó/àozhōu/yīngguó/jiānádà shì duōshǎo qián?)</i></dd>
<dt> Where can I find an Internet cafe?&nbsp;</dt>
<dd> 哪里有网吧? (<i>nǎlǐ yǒu wǎng ba?</i>)</dd>
<dt> How much is it per hour?&nbsp;</dt>
<dd> 一小时是多少钱?  (<i>yī xiǎoshí shì duōshǎo qián?</i>)</dd></dl>

"""
```
